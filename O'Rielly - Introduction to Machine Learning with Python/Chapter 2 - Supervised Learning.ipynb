# Chapter 2: Supervised Learning
Supervised Learning is used whenever we want to predict a cetrain outcome from a given input, and we have examples of input/output pairs, from which we build a ML Model from, comprising out Training Set, to make predictions for new, never-before-seen data. Supervised learning needs human effort to build the training set, but afterwards automates and often speeds up an otherwise laborious or infeasible task. 

## Types of Supervised ML Problems

There are two major types of Supervised Machine Learning problems:
- Classification
- Regression

## Classification
In Classification, our goal is to predict a class label, which is a choice from a predefined list of possibilities.    


#### Types of Classification
- Binary Classification
- Multiclass Classification

#### Binary Classification
Distinguishing between exactly two classes comes under Binary Classification.    
It can be defined as answering a yes/no question.    

Example: Classifying email as either spam or not spam. 

_In binary classification we often speak of one class being the positive class and the other class being the negative class. Here, positive doesn’t represent having benefit or value, but rather what the object of the study is. So, when looking for spam, “positive” could mean the spam class. Which of the two classes is called positive is often a subjective matter, and specific to the domain._

#### Multiclass Classification
Distinguishing between more than two classes comes under Multiclass Classification.

Example: Predicting what language a website is in form the text on the website. The classes here would be predefined list of possible languages. 

### Regression

In Regression, our goal is to predict a Continuous/Floating-Point/Real Number.

Example:
1. Predicting a person's annual income from their education, age and where they life. When predicting income, the predicted value is an amount, and can be any number in a given range. 
2. Predicting yield of corn farm given attributes such as previous yields, weather, no. of employees working on farm. The yield gain can be an arbitrary number.

### Distinguishing between Regression and Classification
A good way of distinguishing between Regression and Classification is continuity. If there is continuity, then it is a Regression problem. Else, it is a Classification Problem. 

Example: 
- On Predicting annual income, there is a clear continuity. Whether a person makes (and similarly, our model predicts) 40,000 or 41,000, there is no tangible difference. Therefore, it is a regression problem. 
- On Predicting species of Iris Flower, there is no matter of degree. It is either of one species, or another. There is no in-between. Therefore, it is a classification problem. 

## Generalisation, Overfitting and Underfitting


### Generalisation
If a model is able to make accurate predictions on unseen data, we say it is able to _generalise_ from the training set to the test set.
We want to make a model that is able to generalise as accurately as possible.

### Overfitting
Overfitting occurs when you fit a model too closely to the training set and obtain a model that works well on the training set but not is able to generalise to new data. In other words, it finds relations and patterns that explains the training set, but are there just by chance and are not representative of real world. 

### Underfitting
Underfitting occurs when the model builds itself from an extremely simple model, which causes low points for our model to learn from, causing lack of knowledge, and it fails in real life. 

## Relation of Model Complexity to Dataset size
Complexity is ultimately tied to the variation of inputs in the training set. Usually, more data points will lead to  more variety, hence larger datasets allow building more complex models. However, there should be more diverse datapoints. Just more of similar data will not help.

## Supervised Machine Learning Algorithms
Some popular supervised machine learning algorithm are:


### k-Nearest Neighbors


Building models consists only of storing the training dataset. To make prediction for a new data point, the algorithm finds the closest data points in the training dataset - it's _nearest neighbours_. 

### k-Neighbours Classification
k-Nearest Neighbors algorithm only considers exactly one nearest neighbour, which is the closest training datapoint to the point we want to make a prediction for. The prediction is then simply the known output for this training point. 

````python
mglearn.plots.plot_knn_classification(n_neighbors=1)
````

Here, 3 new datapoints were added, shown as stars. For each of them, the closest point in the training set was marked. The prediction of the one-nearest neighbours is the label of that point (shown by the color of the cross). 

Instead of considering only the closest neighbor, an arbitraty number k of neighbors can also be considered.    
When considering more than one neighbours, _voting_ is used to assign a label. This means that for each test point, we count how many neighbors belong to class 0 and how many neighbours belong to class 1.    

````python
mglearn.plots.plot_knn_classification(n_neighbors=3)
````

This illustration is for binary classification, but the method can be applied to datasets with any number of classes.    
For more classes, we count how many neighbors belong to each class, andn again predict the most common class.

Let's Investigate whether we can confirm the connection between model complexity and generalisation.    
We will do this on the real-world Breast Cancer dataset.

We will begin by splitting the dataset into a training and test set.

````python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target,
    stratify=cancer.target, random_state=66
)
````

Then we evaluate different test and train accuracy with different numbers of neighbors.

````python
training_accuracy = []
test_accuracy=[]

# Try n_neighbors from 1 to 10
neighbors_settings = range(1,11)

for n_neighbors in neighbors_settings:
    # Build the Model
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    # Record Training Set Accuracy
    training_accuracy.append(clf.score(X_train, y_train))
    # Record Generalisation Accuracy
    test_accuracy.append(clf.score(X_test, y_test))
    
plt.plot(neighbors_settings, training_accuracy, label="Training accuracy")
plt.plot(neighbors_settings, test_accuracy, label="Test accuracy")
plt.xlabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
````

Considering the single nearest neighbor, the prediction on the training set is perfect. But when more neighbors are considered, the model becomes simpler and the training accuracy drops.     
The test set accuracy for using a single neighbor is lower than when using more neighbors. On the other hand, when considering 10 neighbors, the model becomes too simple and performance drops.

The k-Nearest Neighbors algorithm for regression is implemented in the `KNeighborsRegressor` class in `scikit-learn`.

````python
from sklearn.neighbors import KNeighborsRegressor

X, y = mglearn.datasets.make_wave(n_samples=40)

# Split into training and test set:
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# Instantiate Model with 3 Neighbors
reg = KNeighborsRegressor(n_neighbors=3)

# Fit the model using the training data and targets:
reg.fit(X_train, y_train)

# Predict the test set
print(f"""Test Set Predictions:
{reg.predict(X_test)}""")
````

We can now evaluate the model using `score` meathod, which for regressors returns $ R^2 $ score.

````python
print("Test Set R^2: ", reg.score(X_test, y_test))
````

Here, the score is around 0.83, indicating a relatively good model fit.

### Analysing KNeighborsRegressor
For our 1D dataset, we can see what the predictions look like for all possible feature values:


````python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# create 1,000 data points, evenly spaced between -3 and 3
line = np.linspace(-3, 3, 1000).reshape(-1, 1)

for n_neighbors, ax in zip([1, 3, 9], axes):
    # make predictions using 1, 3, or 9 neighbors
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0), markersize=8)
    ax.plot(X_test, y_test, 'v', c=mglearn.cm2(1), markersize=8)

    ax.set_title(
        "{} neighbor(s)\n train score: {:.2f} test score: {:.2f}".format(
            n_neighbors, reg.score(X_train, y_train),
            reg.score(X_test, y_test)))
    ax.set_xlabel("Feature")
    ax.set_ylabel("Target")
axes[0].legend(["Model predictions", "Training data/target",
                "Test data/target"], loc="best")
````

As we can see, using only 1 neighbor, each point on the training set has an obvious influence on the predictions, and the predictions go through all data points, resulting in unsteady predictions.    
    More neiighbors leads to smoother predictions, but these do not fit the training data well

In practice, using a small number of neighbors like three or five often works well, but you should adjust it.

### Strengths of k-NN
- It is easy to understand.
- Gives reasonable performance without a lot of adjustments. 
- Building k-NN model is usually very fast.

### Weaknesses of k_NN
- When training set is very large, predictions can be slow.
- It is important to pre-process your data.

Therefore, k-Neighbors algorithm is easy to understand, it is often not used in practice due to prediction being slow and its inability to handle many features.

## Linear Models
Linear Models make predictions using a _linear function_ of the input features.

## Linear Models for Regression
For regression, the general prediction formula for a linear model looks as follows:    
$$ \hat{y} = w[0] * x[0] + w[1] * x[1] + \dotsb + w[p] * x [p] + b $$    
Here, $ x[0]$ to $ x[p]$ denotes the features. (In this example, the no. of features is $ p + 1$), $w$ and $ b$ are parameters that are learnt, and $ \hat{y}$ is the prediction the model makes.     
For a single feature dataset, this is:    
$$ \hat{y} = w[0] * x[0] + b $$
This is similar to the equation of line ($ y = mx + c $).
Here, $w[0]$ is the slope ($m$) and $b$ is the y-axis offset ($c$).    
The predicted respons can be considered as the weighted sum of input features, with weights (positive or negative) given by entries of $w$.    

````python
mglearn.plots.plot_linear_regression_wave()
````

Looking at $w[0]$, we can see that the slope sould be around 0.4, confirmed visually in the plot. Intercept is where the line sould cross y-axis. This is slightly below 0, as confirmed in the image. 

Linear Models for Regression can be characterised as regressoin models for which the prediction is a line for a single feature, a plane when using two features, a hyperplane in higher dimensions (i.e. when using more features). 

Comparing the predictions to the ones made by KNeighborsRegressor in 'Analysing KNeighbporsRegressor', a straight line looks very restrictive, and looks like all fine details are lost. It is, in a sense, true. It is a strong,  somewhat unrealistic assumption.   
For datasets with many features, linear models can be very powerful. In particular, a case with more features than training datapoints, any target $ y$ can be perfectly modeled as a linear function.

## Ridge Regression
It is also a linear model for regression, so the formula it uses to make predictions is the same one used for ordinary least squares. 
However, the coefficients are chosen not only so that they predict well on the training data, but also fit additionsal constraint.    
The magnitude of coefficients should also be as small as possible (i.e. close to 0). I.e. each feature should have little effect on outcome (translates to having a small slope) while still predicting well. THis constraint is an example of what's called an _regularization_. 

> Regularization measns explicitly restricting model to avoid overfitting.

The particular kind used by Ridge Regression is called L2 Regularization.

````python
from sklearn.linear_model import Ridge

ridge = Ridge().fit(X_train, y_train)         # Data is from earlier example
````

Training Set Score:

````python
ridge.score(X_train, y_train)
````

Test Set Score:

````python
ridge.score(X_test, y_test)
````

As seen, Training score of Ridge is lower than LinearRegression, and Test Score is higher, avoiding overfitting.

Ridge makes a trade-off in training set peroformance for simplicity of model, avoiding overfitting. Importance of the trade-off can be set by the `alpha` parameter. (Default = 1.0)    
Increasing alpha forces ceofficients to be more towards 0, decreaasing training set performance, but might help in generalisation. 

EXAMPLE:

````python
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
print(f"Training set score: {ridge10.score(X_train, y_train)}")
print(f"Test set score: {ridge10.score(X_test, y_test)}")
````

For very small values of alpha, we end up with model ressembling LinearRegression.

````python
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
print(f"Training set score: {ridge01.score(X_train, y_train)}")
print(f"Test set score: {ridge01.score(X_test, y_test)}")
````

The coef_ value is a good indication of effect of change in alpha.

````python
plt.plot(ridge.coef_, 's', label='Ridge alpha=1')
plt.plot(ridge10.coef_, 's', label='Ridge alpha=10')
plt.plot(ridge01.coef_, 's', label='Ridge alpha=0.1')
plt.plot(lr.coef_, 's', label='LinearRegression')

plt.xlabel("Coefficient Index")
plt.ylabel("Coefficient Magnitude")
plt.hlines(0,0,len(lr.coef_))
plt.ylim(-25,25)
plt.legend()
````

Another way to see the effect of regularisation is to fit the value of `alpha` but vary the amount of training data.

## Lasso

Alternative to `Ridge` for regularising LinearRegression, `lasso` also restricts coefficients close to 0, but in a slightly different way, called _L1 Regularisation_.

The consequence of L1 Regularisation is that when using a Lasso, some coefficients are exactly 0, i.e. some features are completely ignored. This can be seen as a form of automatic feature selection.    
Having some coefficients 0 makes model easier to interpret and can reveal most important features of the model. 

````python
from sklearn.linear_model import Lasso

lasso = Lasso().fit(X_train, y_train)
print('Training set score: ', lasso.score(X_train, y_train))
print('Test set score: ', lasso.score(X_test, y_test))
print('Features used :', np.sum(lasso.coef_!=0))
````

Bad performance shows underfitting.     
It only used 3 of the 105 features.

Lasso also uses an `alpha` parameter.

Using the same alpha as previous to 1.0, we get:

````python
# We increase the default value of 'max_iter', or the model
# will throw warnings to do so.
lasso001 = Lasso(alpha=0.01, max_iter=100000).fit(X_train, y_train)
print("Training set score: ", lasso001.score(X_train, y_train))
print("Training set score: ", lasso001.score(X_test, y_test))
print("Number of features used: ", np.sum(lasso001.coef_ != 0))
````

Lower alpha allowed more complex model fitting, which worked better on the training and test data.

Too low `alpha` removes the effect of regularisation and causes overfitting, with result similar to `LinearRegression`. 

````python
lasso000001 = Lasso(alpha=0.00001, max_iter=1000000).fit(X_train, y_train)

print('Training set score: ', lasso000001.score(X_train, y_train))
print('Test set score: ', lasso000001.score(X_test, y_test))
print('Features used :', np.sum(lasso000001.coef_!=0))
````

````python
plt.figure(figsize=(10, 7))

plt.plot(lasso.coef_, 's', label='Lasso alpha=1')
plt.plot(lasso001.coef_, '^', label='Lasso alpha=0.01')
plt.plot(lasso000001.coef_, 'v', label='Lasso alpha=0.00001')

plt.plot(ridge01.coef_, 'o', label='Ridge alpha=0.1')
plt.legend(ncol=2, loc=(0, 1.05))
plt.grid()
plt.ylim(-25, 25)
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitute")
````

- For `alpha=1` (Shown as squares), most of the coefficients are either 0 or small in magnitude. 
- Decreasing `alpha=0.01` (Shown as Upward Triangle) causes most features to be exactly 0. 
- Using `alpha=0.00001` (Shown as Downward triangles), we get quite an unregularised model, with most coefficients being non zero with high magnitude. 
- The `Ridge` solution with `alpha=0.1` has similar performance to `Lasso` with `alpha=0.01` , but using `Ridge`, all coefficients are non-zero. 

`Ridge` Regression is generally the first choice, but with large amounts of features and only expect a few of them to be important, `Lasso` might be better choice.

`Lasso` is also easier to interpret and understand, as it is only a subset of the input features.

`scikit-learn` also provides the `ElasticNet` class, which combines the penalties of `Lasso` and `Ridge`.    <br>
In practice, this combination works best, but needs adjustments to 2 parameters (one for L1 Regularisation and other for L2 Regularisation).

## Linear Models for Classification
Linear Models are also used for classification.

In case of Binary Classification, prediction is made using the following formula:

$$ \hat{y} = w[0] * x[0] + w[1] * x[1] \dotsc + w[p] * x[p] + b > 0 $$ 

Formula is similar to Linear Regression, but instead of weigthed sum of features, we threshold the predicted value to 0.

If Predicted vale is < 0, we predict class -1, if it is > 0, we predict class 1.

Fir linear function of classification, the _decision boundary_ is the linear function of input. i.e., a (binary) linear classifier seperates two classes using a line, a plane or a hyperplane.

There are many algorithms for linear models, that differ in two ways:
- The way in which they measure how well a particular combinations of coefficients and itercepts fits the training data.
- What kind of regularisation they use.

For technical mathematical reasons, it is not possible to adjust `w` and `b` to minimize the number of misclassifications the algorithm makes. For most purposes, the different choices for item 1 in the preceding list (called loss functions) are of little significance. 

Two most common linear classification algorithms are:
- **Logistic Regresion** implemented in `linear_model.LogisticRegression`. (Despite it's name, it is a classification algorithm and not a regression algorithm)
- **Linear Support Vector Machines** implemented in `svm.LinearSVC` (SVC stands for Support Vector Classifier)

````python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
````

````python
# Make data:
X, y = mglearn.datasets.make_forge()

fig, axes = plt.subplots(1,2,figsize=(10,3))

for model, ax in zip([LinearSVC(max_iter=3524), LogisticRegression()], axes):
    ## Max Iter was specified manually, to supress the warning. Ignore it...
    clf = model.fit(X,y)
    
    mglearn.plots.plot_2d_separator(clf, X, fill=False, eps=0.5,ax=ax, alpha=0.7)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title(f"{clf.__class__.__name__}")
    ax.set_xlabel("Feature 0")
    ax.set_xlabel("Feature 1")
axes[0].legend()

````

By default, both models apply L2 Regularisation.

For `LogisticRegression` and `LinearSVC` the trade-off parameter that determines the strength of the regularization is `C`.<br>
Higher the `C`, less the regularisation. 

Another interesting aspect of how `C` acts, low values cause the algrithms to try and adjust majority of datapoints, and higher stresses the importance that each individual datapoint be classified correctly.

## Linear Models for Multiclass Classification
Many Linear Classification models are for Binary Classificaion only, and don't extend naturally to multiclass case (exception being LogisticRegression).

A common technique to multicalss classification is `one-vs-rest` approach. In `one-vs-rest` approach, a binary-model is learnt for each class that tires to seperate that class from all of the other classes, resulting in as many binary models as there are classes.<br>
To make a prediction, all binary classifiers are run on a test point. The classifier that has the highest score on its single class 'wins' and this class labeled is returned as prediction. 

Having one binary classifier per class results in having one vector of coefficients (w) and one intercept (b) for each class.The class for which the result of the classification confidence formula given here is the highest is the assigned class label:
$$ w[0] * x[0] + w[1] * x[1] + \dotsc + w[p] * x[p] + b $$

The mathematics mehind multiclass LogisticRegression is somewhat different, but they also result in one coefficient vector and intercept per class, and the same method of making a prediction is applied.

Applying `one-vs-rest` method to a simple three-class classification dataset. We use a two-dimensional dataset, where each class is given by data sampled from a Gaussian Distribution.

## Naive Bayes Classifier

Native Bayes Classifier are similar to Linear Models, but tend to be faster in training in exchange for loss of generalisation performance. 
Their efficiency is due to looking at each feature indivisually and collecting simple per-class statistics from each feature.

There are three kinds of Naive Bayes Classifiers implemented in `sklearn`:
- `GaussianNB`
- `BernoulliNB`
- `MultinomialNB`

`GaussianNB` can be applied to any continueous data, while `BernoulliNB` assumes Binary Data, and `MultinomialNB`assumes count data.

`BernoulliNB` and `MultinomialNB` are mostly used in text data classification. 

`BernoulliNB` counts how often every feature of each class is not zero.
Example:

````python
x = np.array([[0,1,0,1],
             [1,0,1,1],
             [0,0,0,1],
             [1,0,1,0],])
y = np.array([0,1,0,1])
````

We have 4 datapoints with 4 binary features each. There are 2 classes, 0 and 1. For class 0 (1st and 3rd datapoints), the first feature is 0 two times and non-0 zero times, the second feature is 0 one time and non-0 one time and so on. These same counts are then calculated for datapoints in the second class. Counting the non-zero entries per class in essence looks like this:

````python
counts = {}

for label in np.unique(y):
    # Iterate over each class
    # count(sum) enteries of 1 per feature
    counts[label] = X[y==label].sum(axis=0)
print("Feature counts:\n{}".format(counts))
````

The other two Naive Bayes classifier are slightly different in what kinds of statistics they compute. `MultinomialNB` takes into account the average value of each feature for each class, while `GaussianNB` store the average value as well as the standard deviation of each feature for each class.

To make a prediction, a datapoint is compared to the statistics for each of the class, and the best matching class is predicted. 
> Interestingly, for both MultinomialNB and BernoulliNB, this leads to a prediction formula that is of the same form as in the linear models.<br>

The `coef_` for naive Bayes models has a somewhat different meaning than in the linear models, in that the `coef_` is not same as `w`.

### Strengths, Weaknesses and Parameters
`MultinomialNB `and `BernoulliNB` have a single parameter`alpha` which controls complexity.<br>
`alpha` works by the model adding to the data `alpha` many virtual data points that have a positive value for all features, resulting in smoothing of all statistics.<br>
Large `alpha` means more smoothing, resulting in less complex models. Model is robust to setting of `alpha`, meaning that setting `alpha` is not critical to good performance. However, tuning it usually improves accuracy somewhat.<br>


`GaussianNB` is mostly used on a very high-dimensional data, while the other two variants of naive Bayes are widely used for sparse count data such as text. `MultinomialNB` usually performs better than `BernoullieNB` , particularly on datasets with many non-zero features (i.e. Large Documents).

Naive Bayes share many of the strengths and weaknesses of linear models, and are used often as a baseline models and on large datasets, where even a linear model might take too long. 

## Decision Trees

Decision Trees are widely used models for clasification and regression tasks.      
They essentially learn a hierarchy of if/else questions, leading to a decision.

EXAMPLE:

Each node in the tree either represents a question or a terminal node (aka Leaf) that contains the answer. Edges connect the answers to a question with the next question you would ask.

### Building Decision Trees
Building Decison Tree for the 2D Classification Dataset `two_moons`:


Learning a decision tree means learning the sequence of if/else questions that gets us to the true answer most quickly. These questions are called *tests* (not to be confused with _test set_). Generally, data does not come in form of binary yes/no features as in the animal example, but is instead represented as continueous features such as in the 2D Datasets shown in figure above.

To build a tree tree, the algorithm searches over all possible tests and finds one that is most informative about the target variable.

First ten picked:


Splitting the dataset horizantally at $ x[1] = 0.0596$ yields most information, and best seperates class 0 from points in class 1.
The _root node_ represents the whole dataset, consisting of 75 points belonging to class 0 and 75 points belonging to class 1.

The split is done by testing whether $x[1] <= 0.0596$ indicated by a black line. If the test is true, a point is assigned to the left node, containing 2 points belonging to 0 and 32 to 1. Otherwise, the point is assigned to right node, containing 48 points belonging to class 0 and 18 points to class 1. These two nodes correspond to the top and bottom regions shown in above figure. 

We can build a more accurate model by repeating to process of looking for best test in both regions:


This recursive process yields a binary tree of decisions, with each node containing a set, the alogorithm effectively building a hierarchial pattern. As each test only concerns a single feature, the regions in the resulting partition always has axis-parallel boundaries.

The recursive partitioning is done until each region only contains single target value (Class or Regression).

Leaf containing homogeneous datapoints is called *pure*

Prediction is made by checking the region of the partition of the feature space the point lies in, and then predicting the majority target (or the single target in case of pure leaves) in that region. 

The region can be found by traversing tree from root and going left or right depending whether the test is fulfilled or not. 

Regression is possible using same technique. To make a prediction, we traverse the tree to test each node and find the leaf the new datapoint falls into.  The output of this datapoint is the mean target of training point. 

### Controlling Complexity of Decision Trees
Building a tree as described and continuing until all leaves are pureleads to extremely high complexity and overfit.    
There are two common strategies to prevent overfitting:
- Stopping the creation of the tree early
- Building the tree but then removing or collapsing nodes that contain little information (aka **Post-Pruining**, or just **Pruining**)

Decision trees in `sklearn` are implemented in the `DecisionTreeRegressor` and `DecisionTreeClassifier` classes.    
`sklearn` only implements pre-pruining, and not post-pruining.

Looking at effect of pre-pruining on the Breast Cancer dataset:

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target)
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)
print(f"Accuracy on Training Set: {tree.score(X_train, y_train)}")
print(f"Accuracy on Test Set: {tree.score(X_test, y_test)}")
````

The data is overfitting...

We have to restrict the depth of a decision tree, the tree can become arbitrarily deep and complex. Unpruned trees are therefore prone to overfitting and not generalising well to new data. 

Applying pre-pruining to the tree, we will stop developing the tree before we perfectly fit to the training data. One option is to stop building the tree after a certain depth is reached.    
This can be set with the `max_depth` parameter.

````python
tree = DecisionTreeClassifier(max_depth=4, random_state=0)
tree.fit(X_train, y_train)

print(f"Accuracy on Training Set: {tree.score(X_train, y_train)}")
print(f"Accuracy on Test Set: {tree.score(X_test, y_test)}")
````

Setting a limit using `max_depth=4` (which basically means limiting the consecutive questions to 4), we get worse accuracy on training set, but better accuracy on test set, preventing overfitting.
